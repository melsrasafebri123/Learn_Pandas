```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"/content/LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"/content/LOTR 2.csv")
df2
```

```python
#Inner join
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
#Full outer join
df1.merge(df2, how='outer')
```

```python
#Left join
df1.merge(df2, how='left')
```

```python
#Right join
df1.merge(df2, how='right')
```

```python
#Match one by one row (many row=nxm)
df1.merge(df2, how='cross')
```

```python
#Similar to a merge but it's better if both tables are indexed [example is below]
df1.join(df2, on='FellowshipID', how='outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
#Use join when both tables are set index
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
#Join but table 1 first then table 2, like sum
pd.concat([df1,df2])
```

```python
#Merge but only show the same column
pd.concat([df1,df2],join='inner')
```

```python
#Merge not down but side by side
pd.concat([df1,df2],join='outer',axis=1)
```

```python
#Not recommend beccause there's warning
df1.append(df2)
```
